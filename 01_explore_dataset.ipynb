# 📡 TelecomRAG – Dataset Exploration
This notebook walks you through the TeleQnA dataset before ingesting it.


```python
import sys, json
sys.path.insert(0, '..')

from src.data_loader import download_teleqna
import pandas as pd

json_path = download_teleqna()
with open(json_path) as f:
    data = json.load(f)

print(f'Total questions: {len(data)}')
```

```python
# Peek at one entry
first_key = list(data.keys())[0]
data[first_key]
```

```python
# Category distribution
df = pd.DataFrame(data.values())
df['category'].value_counts().plot(kind='bar', figsize=(10,4), title='Questions per category')
```

```python
# Answer distribution (how often is each option correct?)
df['answer'].value_counts()
```

```python
# Explanation length
df['explanation_len'] = df['explanation'].fillna('').apply(len)
df['explanation_len'].hist(bins=30, figsize=(8,3))
```
